# Import our images for RGB assignment

```python
# f125 will be gree
g_name = get_pkg_data_filename(r'C:\Users\Pablo\Downloads\hlsp_hudf12_hst_wfc3ir_udfmain_f125w_v1.0_drz.fits')
# f160 is red
r_name = get_pkg_data_filename(r"C:\Users\Pablo\Downloads\hlsp_hudf12_hst_wfc3ir_udfmain_f160w_v1.0_drz.fits")
#105 blue
i_name = get_pkg_data_filename(r"C:\Users\Pablo\Downloads\hlsp_hudf12_hst_wfc3ir_udfmain_f105w_v1.0_drz.fits")
#read in files with astropy functions
g = fits.open(g_name)[0].data
r = fits.open(r_name)[0].data
i = fits.open(i_name)[0].data
#defint the mean of the images data for out boundary conditions
mean = np.mean(g+r+i)
rgb_default = make_lupton_rgb(r,g,i, Q=10, stretch=0.01, filename="hud.png")
#print image and define upper and lower bounds as one standard deviation from mean to display best image
plt.imshow(rgb_default, interpolation='nearest', vmin= mean-np.std(g+r+i), vmax= mean+np.std(g+r+i), origin='lower')
```
